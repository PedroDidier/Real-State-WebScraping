## Importing Libs

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from pandas_profiling import ProfileReport
%matplotlib inline
```

## Importing EDA DataSet

```python
df = pd.read_csv('../DataSets/EDA_OLX_housing.csv')
```

# Pandas Profiling Report - 1

Pandas profiling report does A LOT for us automatically. It saves a .hltm with a really good insight for what needs to be focused in the dataset. Despite this being a powerful tool, it is always preferable to complement its analysis with hand-made plots on Seaborn and MatplotLib.

```python
prof = ProfileReport(df)
prof.to_file(output_file='EDA_OLX_housing_PandasProf.html')
```

### Conclusions

- Most of data is related to houses for sale, missing data should not be considered
- Rooms has a lot of missing data, we can input that with no worries
- Area has dangerous outliars, which should be deleted
- Car Spots has most of its data missing, we should input 0 for those
- Condominium Fee also have bad outliars, which should be deleted
- Price its (obviously) very influenced by "Status", ML analysis should focus only one of this categories
- There are duplicate rows

### Cleaning Data after first report

```python
# Conclusion n1
df=df[df['Status']=='For sale']
```

```python
# Conclusion n2 will be dealt in pre-processing
```

```python
# Conclusion n3
df=df[df['Area (m²)']<100000] # 10 hectars is the maximum accepted
```

```python
# Conclusion n4
df['# Car Spots']=df['# Car Spots'].replace(np.nan,0)
```

```python
# Conclusion n5
df=df[df['Condominium Fee (BRL)']<10000]
df['Condominium Fee (BRL)']=df['Condominium Fee (BRL)'].replace(np.nan,0)
```

```python
# Conclusion n6
del df['Status']
```

```python
# Conclusion n7
df = df.drop_duplicates()
```

```python
# these values can not be 0
df = df[df['Price (BRL)']>0]
df = df[df['Area (m²)']>0]
```

## Square Meter Price

Price per m² is a very discussed metric in Real State conversations, let's see how it works on the data we have colected

```python
df['Price per m²'] = df['Price (BRL)'] / df['Area (m²)']
```

```python
sns.displot(df['Price per m²'])
```

## !BIG ISSUE!

Values above 20000 reais per m² are simply not valid. Let's discard those before we continue

```python
df = df[df['Price per m²']<20000]
```

```python
sns.displot(df['Price per m²'])
```

## Lets save the dataset

```python
# reseting index
df = df.reset_index(drop=True)
```

```python
# saving changes
df.to_csv('../DataSets/EDA_OLX_housing_AfterProf.csv',index_label=False)
```

# Other plots

Pandas Profiling does most of the stuff for us, but, when it comes to specific plots, we need to do it ourselves. I like to plot some things without outliars as well so we have a more general ideia of the real relations between columns.

```python
sns.displot(df[df['Price (BRL)']<df['Price (BRL)'].mean()]['Price (BRL)'])
```

This is the Price distribution considering only values below the mean of the column, since, as we can see in the profiling, there is a low sample above the mean price

```python
sns.lmplot(x='Area (m²)',y='Price (BRL)',data=df)
```

There seems to be something close to a linear correlation between "Price" and "Area", as we can also notice in the hexagonal jointplot from the Pandas Profiling

```python
sns.lmplot(x='Condominium Fee (BRL)',y='Price (BRL)',data=df)
```

There also seems to be something close to a linear correlation between "Condominium Fee" and "Price", as we can also notice in the hexagonal jointplot from the Pandas Profiling

## Recife

Recife is the capital of Pernambuco and greatest city of the RMR, we will differ it from the rest and re-do some analysis

```python
def is_recife(city_name):
    if city_name == 'Recife':
        return 'Recife'
    return 'RMR'
```

```python
df['Recife'] = df['City'].apply(is_recife)
```

```python
sns.pairplot(df,hue='Recife',palette='coolwarm')
```

Conclusions:
- City == Recife has more instances than any other city, as expected
- Recife is more expansive in both Condominium Fee and Price (as expected)
